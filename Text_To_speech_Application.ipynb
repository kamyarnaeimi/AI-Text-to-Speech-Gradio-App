```python
import gradio as gr
from openai import OpenAI

client = OpenAI(
    base_url = "https://api.avalai.ir/v1",
    api_key = "aa-tOgDydcQVuRDuSOgUmY1sEzdkt4P0UmgOWfb5jpEttb9wufW"
)

#function to generate speech from text

def text_to_speech (text):
    speech_file_path = "output_ttc.mp3" #output file
    try:
        #generate speech using AvalAI API
        response = client.audio.speech.create(
            model = "tts-1",
            voice = "alloy",
            input = text,
        )
        response.stream_to_file(speech_file_path) #save and generate speech to a file
        return speech_file_path # returning the file path for playback and download
    except Exception as e:
        return f"Error: {str(e)}"

#Gradio AI

iface= gr.Interface(
    fn=text_to_speech,
    inputs=gr.Textbox(label="Enter your text here"),
    outputs=gr.Audio(label= "Generated Speech"),
    title = "AI Powered (TTS)",
    description = "Enter any text and an AI-generated voice output"
)
iface.launch()
```
